```python
from tensorflow.keras.applications.vgg16 import VGG16, preprocess_input, decode_predictions
from tensorflow.keras.preprocessing import image
import numpy as np
import matplotlib.pyplot as plt
```

```python
print("---Step 1: Loading the Pre-Trained 'Brain'----")
model = VGG16(weights = 'imagenet')
print("VGG16 Model Loaded Successfully.")
```

```python
print("\n--- Step 2: Preparing Our Custom Image---")
img_path = '/DocScanner Feb 9, 2026 6-24 PM.jpg'
img = image.load_img(img_path, target_size = (224, 224))
x = image.img_to_array(img)
x = np.expand_dims(x, axis = 0)
x = preprocess_input(x)
print("Image Converted To Mathematical Array and Preprocessed.")
```

```python
print("\n---AI Inference (Making Predictions)---")
print("analyzing Visual Data.......")
predictions = model.predict(x)
results = decode_predictions(predictions, top = 3)[0]
```

```python
print("\n ---AI Classification Results ---")
for result in results:
  label = result[1]
  confidence = result[2] * 100
  print(f"Prediction: {label} (Confidence: {confidence:.2f}%)")
  plt.imshow(image.load_img(img_path))
  plt.axis('off')
  plt.show()
```
