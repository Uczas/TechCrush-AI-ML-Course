```python
# TechCrush AI Bootcamp - Week 7, Day 2
# Project: Designing a Convolutional Neural Network (CNN) Architecture
import tensorflow as tf
from tensorflow.keras.models import Sequential
from tensorflow.keras.layers import Conv2D, MaxPooling2D, Flatten, Dense
print("--- Constructing the CNN Architecture ---")
model = Sequential([
# ---------------------------------------------------
# 1. THE EYES (Visual Feature Extraction)
# ---------------------------------------------------
# Layer 1: Slides 32 separate 3x3 filters across a 64x64 pixel color image
# (The '3' indicates Red, Green, Blue color channels)
Conv2D(filters=32, kernel_size=(3, 3), activation='relu', input_shape=(64, 64, 3)),
                     # Layer 2: Shrinks the image by taking the maximum value in a 2x2 grid
MaxPooling2D(pool_size=(2, 2)),

# Layer 3: A second set of "eyes" to look for more complex textures
Conv2D(filters=64, kernel_size=(3, 3), activation='relu'),
MaxPooling2D(pool_size=(2, 2)),
# Layer 4: A third layer for highly specific object features
Conv2D(filters=128, kernel_size=(3, 3), activation='relu'),
MaxPooling2D(pool_size=(2, 2)),
# ---------------------------------------------------
# 2. THE BRAIN (Final Classification)
# ---------------------------------------------------
# Layer 5: We flatten the extracted visual features into a 1D array
Flatten(),
# Layer 6: A standard hidden layer to process the visual findings
Dense(128, activation='relu'),
# Layer 7: Output layer. Let's assume we are classifying 3 specific things
# (e.g., Clear Water, Algae Bloom, Floating Debris)
Dense(3, activation='softmax')
])
print("\n--- Network Diagnostics ---")
# This command prints a table showing exactly what happens to the image at each layer
model.summary()
```
